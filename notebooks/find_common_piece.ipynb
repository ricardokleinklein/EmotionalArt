```python
import numpy
import matplotlib.pyplot as plt
emotion_names = ["Amusement", "Awe", "Contentment", "Excitement",
                 "Disgust", "Anger",
                 "Fear", "Sadness", "Something Else"]
emotion_names = [*emotion_names, emotion_names[0]]
#emotion_example = [0, 3, 2, 7, 0, 0, 0, 0]
emotion_example = [3, 0, 1, 1, 0, 0, 0, 0, 1]
emotion_example = [*emotion_example, emotion_example[0]]
label_loc = numpy.linspace(start=0, stop=2*numpy.pi, num=len(emotion_names))
plt.figure(figsize=(8, 8))
plt.subplot(polar=True)
plt.plot(label_loc, emotion_example)
lines, labels = plt.thetagrids(numpy.degrees(label_loc),
                               labels=emotion_names, fontsize=20)
```

```python
import plotly.graph_objects as go
import plotly.offline as pyo

fig = go.Figure(
    data=[
        go.Scatterpolar(r=emotion_example, theta=emotion_names, fill="toself")
    ],
    layout=go.Layout(
        title=go.layout.Title(text=""),
        polar={'radialaxis': {'visible': True}},
        showlegend=True,
        template="presentation"
    )
)
fig.update_layout(
    font=dict(
        family="Times New Roman",
        size=50
    )
)
pyo.plot(fig)
```
